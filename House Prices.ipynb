More informations about concatenate data:

https://pandas.pydata.org/docs/reference/api/pandas.concat.html

```python
all_data = pd.concat([df, df2], axis=0, sort=False)
all_data = all_data.drop(['Id', 'SalePrice'], axis=1)
```

Looking for NULL values:

```python
Total = all_data.isnull().sum().sort_values(ascending=False)
percent = (all_data.isnull().sum() / all_data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([Total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(25)
```

Delete those values:

```python
all_data.drop((missing_data[missing_data['Total'] > 5]).index, axis=1, inplace=True)
print(all_data.isnull().sum().max())
```

```python
Total = all_data.isnull().sum().sort_values(ascending=False)
Total.head(30)
```

List dataframe with the columns of all_data (concatenate dataframe):

```python
Total_from_all_data = all_data.loc[:,Total.index]
```

```python
Total_from_all_data
```

List only dtype = 'float':

```python
Total_from_all_data.loc[:,all_data.dtypes == 'float']
```

Creating list for these specific columns (floats with missing values):

```python
isnull_float = list(Total_from_all_data.loc[:,all_data.dtypes == 'float'].columns)
```

```python
isnull_float
```

More informations about filling NULL:

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html

```python
for val in isnull_float:
    all_data[val] = all_data[val].fillna(0)
```

Checking again Total:

```python
Total_wf = all_data.isnull().sum().sort_values(ascending=False)
Total_wf.head(30)
```

Doing the same for objects:

```python
Total_from_all_data.loc[:,all_data.dtypes == 'object']
```

```python
isnull_object = list(Total_from_all_data.loc[:,all_data.dtypes == 'object'].columns)
```

```python
isnull_object
```

More informations about filling NULL for objects:

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.mode.html

```python
for val in isnull_object:
    all_data[val] = all_data[val].fillna(all_data[val].mode()[0])
```

And checking again Total:

```python
Total_wf_wo = all_data.isnull().sum().sort_values(ascending=False)
Total_wf_wo.head(30)
```

```python
all_data.isnull().sum().max()
```

More informations about skew (including an interesting article on towardsdatascience):

https://towardsdatascience.com/top-3-methods-for-handling-skewed-data-1334e0debf45

https://numpy.org/doc/stable/reference/generated/numpy.log1p.html

```python
non_objects = all_data.dtypes[all_data.dtypes != 'object'].index
```

```python
skews = all_data[non_objects].skew().sort_values(ascending = False)
```

```python
skews
```

Taking only the skews high enough:

```python
skews_high = skews[abs(skews) > 1]
```

```python
skews_high
```

```python
import numpy as np

for val in skews_high.index:
    all_data[val] = np.log1p(all_data[val])
```

Checking again the skews:

```python
skews_log = all_data[non_objects].skew().sort_values(ascending = False)
```

```python
skews_log
```

Getting dummies:

```python
all_data = pd.get_dummies(all_data)
all_data.head()
```

Preparation for training:

```python
y_train = df['SalePrice']
```

```python
x_train =all_data[:len(y_train)]
x_test = all_data[len(y_train):]
```

Checking the shapes:

```python
print(x_train.shape)
print(x_test.shape)
```

More informations about RandomForestRegressor:
    
https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, max_depth=10)
model.fit(x_train, y_train)
```

```python
predicted_y = model.predict(x_test)
```

```python
predicted_y
```

```python
test_id = df2['Id']

sub = pd.DataFrame()
sub['Id'] = test_id
sub['SalePrice'] = predicted_y
sub.to_csv('mysubmission.csv',index=False)
```
